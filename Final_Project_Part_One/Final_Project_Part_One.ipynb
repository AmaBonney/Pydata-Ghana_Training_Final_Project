```python
from bs4 import BeautifulSoup as soup
from urllib.request import Request, urlopen
import pandas as pd
```

```python
data = {}
```

## Scraping Details for Natura XL

```python
url = "https://www.yarnplaza.com/product/4216/dmc-natura-xl.html?sqr=natura%20xl"
headers = {
    "User-Agent": "Mozilla/6.0"
}
```

```python
request = Request(url, headers=headers)
web_page = urlopen(request).read()
web_page
```

```python
page = soup(web_page, "html.parser")
page.prettify()
```

```python
# Brand Name
table = page.find("div", id="pdetailTableSpecs")
brand_name = table.find_all("tr")[17].text[8:]
needle_size = table.find_all("tr")[4].text[11:]
composition = table.find_all("tr")[3].text[11:]
product_name = page.find("h1", id="pageheadertitle").text[4:13]
product_price = page.find_all("span", class_="product-price")[0]['content']
product_price = str(round(float(product_price)))
```

```python
data["Brand"] = [brand_name]
data["Name"] = [product_name]
data["Price()"] = [product_price]
data["Needle Size"] = [needle_size]
data["Composition"] = [composition]
data["Delivery Time"] = [""]
data
```
